## 決定木

この節では**決定木（Decision Tree）**によるモデル構築方法を学ぶ．
決定木は，ある目的に到達するためにデータの各属性の条件分岐を繰り返してクラス分けする方法である．
目的変数がカテゴリの場合は**分類木**，数値の場合は**回帰木**とよぶ．

### データの整理

以下では話を簡単にするため，説明変数を`gill_color`（ひだの色），`gill_attachment`（ひだの付き方），`odor`（匂い），`cap_color`（かさの色）の4つに限定することにしよう．
これらのデータは，上記表示の通り，たとえば`gill_color`は`black`のときはk，`brown`のときはnといったカテゴリ変数となっている．
決定木で扱う変数は，説明変数，目的変数，ともに数値型でなければならない．
そのため，このようなカテゴリ変数は数値変数に変換する必要がある．

そこでカテゴリ変数をダミー変数化する．
ダミー変数化するというのは，たとえば性別変数の列に`male`か`female`の値が入っている場合，性別の列を`male`列と`female`列の2列に分けて表現することをいう．
より具合的には，性別の値が`male`であった場合は`male`列を1，`female`列を0にすることである．
他に`one-hot`化する，`one-hot`エンコーディングを施すなどともいう．
Pandasの`get_dummies`関数を用いると，ダミー変数化できる．

```python
mushroom_dummy = pd.get_dummies (mushroom[['gill_color', 'gill_attachment','odor','cap_color']]) 
mushroom_dummy.head(10).T
```

上記の通り変換後のデータは，元の変数名と値の組み合わせになる．
たとえば，`gill_color_k`に1が立っていれば，`gill_color`が`k`であったことを意味する．
ダミー変数化はこのようにカテゴリ変数をフラグ化（数量化）したいときに使える最もシンプルな方法である．

カテゴリを表す目的変数であっても，入力データ形式が数値である必要があるため，目的変数である`classes`についても新しい変数`flg`に変換しておく．
この処理は`classes`変数の値が`p`の場合は1，そうでない場合は0として（`lambda`関数の部分），新しい変数`flg`を追加している．
そして`map`関数を使うことでその処理をすべての要素（セル）に適用している．
ここまでで目的変数を0/1の数値型で表現し直し，カテゴリ変数の特徴量もダミー変数化した．
これで，決定木（アルゴリズム）に入力することができるようになった．

```python
mushroom_dummy['flg'] = mushroom['classes'].map (lambda x: 1 if x =='p' else 0) 
```

### エントロピー：不純度の指標

決定木のモデル構築の前に，決定木の作られ方をカテゴリ識別の不純度（`impurity`）という視点から見る．
不純度とは，毒キノコか否かの識別の状態を表す指標であり，不純度が高いとはカテゴリ識別ができていない状態を意味する．
たとえば，`cap_color`が`c`であるかそうでないかの`TRUE`（1） or `FALSE`（0）でデータを分けるとして，その時にそれぞれ毒キノコがどれくらいあるのかをクロス集計する．
下記に行が`cap_color`がcであるか（1），そうでないか（0），列が毒フラグ`flg`が立っているか（1），そうでないか（0）のクロス集計結果を示す．

```python
mushroom_dummy.groupby(['cap_color_c', 'flg'])['flg'].count().unstack() 
```

上表より，`cap_color`が`c`（`cap_color_c`が1）であれば，毒（`flg`が1）の数が12個で，毒でない（`flg`が0）数が32個であることがわかる．
そして`cap_color`が`c`でなければ（`cap_color_c`が0），毒（`flg`が1）の数が3904個で，毒でない（`flg`が0）数が4176個とわかる．

この結果をみると，どちらを選んでも毒キノコが一定の割合で含まれていため`cap_color`が`c`であるか否かの情報は，毒キノコを見分けるのにあまり役に立たないように見える．

一方，別の変数`gill_color`が`b`であるかそうでないかの`TRUE`（1） or `FALSE`（0）で分けた場合のクロス集計結果は以下のようになる．

```python
mushroom_dummy.groupby (['gill_color_b', 'flg'])['flg'].count().unstack()
```

上表より，`gill_color`が`b`（`gill_color_b`が1）であれば，毒（`flg`が1）の数が1728個で，毒でない（`flg`が0）数が0個（ないので`NaN`）であることが分かる．
そして`gill_color`が`b`でなければ（`gill_color_b`が0），毒（`flg`が1）の数が2188個で，毒でない（`flg`が0）数が4208個とわかる．

先程の分岐条件と比べると，`gill_color`が`b`であるか否かの分岐条件の方が，識別能力の高い（不純度の低い識別状態を導く）有益な条件だとわかる．

ここでは2つの変数の例（`cap_color_c`と`gill_color_b`）で考えた．
しかし，他にもさまざまな変数があり，それぞれに対して上のような条件分岐を考えることができる．
このように決定木とは，多数の変数の中でどの変数から最も有益な条件分岐を得られるかを見分けてくれるアルゴリズムであり，その分岐条件の優劣を決める際に不純度が使われている．
そして，その不純度の指標としてよく使われるものに**エントロピー（entropy）**がある．
エントロピーの定義は以下の式$H(S)$で与えられる．
$S$はデータの集合，$n$はカテゴリの数，$p$は各カテゴリに属するデータサンプルの割合を示す．

\begin{eqnarray}
\ H(S)= -\sum^n_{i=1}(p_i\log_{2}p_i)
\end{eqnarray}

今回の例ではカテゴリは2つ（毒キノコか否か）で，毒キノコでない割合が$p1$，毒キノコである割合が$p2$である．
ここで1つ目の例として，ある分岐条件によって毒キノコも食用キノコも等しい割合で入っている状態を考える．
$p1=p2=0.5$となるので，エントロピーは上の式から以下のように計算できる．
なお，底が2のログ関数（`np.log2`）を使っている．

```python
-(0.5*np.log2(0.5) + 0.5 * np.log2(0.5))
```

上記より，エントロピーが1.0になることが確認された．
情報量の定義より二者択一の予測において，データとしての乱雑さが最大となる場合，エントロピーは1.0となる．

下記のように，たとえば三択であれば約1.585，四択であれば2.0が最大となる．

```python
-((1/3) * np.log2(1/3) + (1/3)*np.log2(1/3) + (1/3) * np.log2(1/3))
```

```python
-(np.log2(1/3))
```

```python
-((1/4) * np.log2(1/4) + (1/4) * np.log2(1/4)+ (1/4)* np.log2(1/4)+ (1/4) * np.log2(1/4))
```

```python
-(np.log2(1/4))
```

この場合，毒キノコもそうでないキノコも等しい割合（0.5）で含まれているため，全く識別ができないことを意味する．
次に，毒キノコでない割合`p1=0.001`，毒キノコである割合が`p2=0.999`の場合を考えてみる．

```python
-(0.001 * np.log2(0.001) + 0.999 * np.log2(0.999))
```

上記の通り，エントロピーは0に近い値になっているのがわかる．
この状態は，ほぼデータは毒キノコと特定できているためエンロトピーは小さくなっている．
まとめると，エントロピーは1.0に近いと識別ができていない状態，0.0に近ければ識別がよくできている状態と言える．
なお，今回の例では，カテゴリは2分類のため$p1 = 1 - p2$という関係式ができるめ，エントロピーの式は以下のように表せる．

```python
def calc_entropy(p):
    return - (p * np.log2(p) + (1-p)*np.log2(1-p))
```

`p`は確率で0から1までの値を取るため，この`p`とエントロピーの式をグラフで表すと以下のようになる．エントロピーは最大で1，最小で0となることが確認できる．

```python
p = np.arange(0.001,0.999,0.01)

plt.plot(p,calc_entropy(p))
plt.xlabel("prob")
plt.ylabel("entropy")
plt.grid(True)
```

ここまでの説明で，エントロピーが識別の不純度を表すということを説明した．
先ほどのキノコのデータでエントロピーを計算してみよう．
扱っているデータセットは合計で8124行あり，目的変数`flg`がカテゴリを表すため，そのデータをカウントする．

```python
mushroom_dummy.groupby('flg')["flg"].count()
```

上記より，毒でないキノコ（0）は4208個，毒キノコ（1）は3916個とわかる．
よって毒キノコでない割合は0.518（=4208/8124），毒キノコである割合は0.482（=3916/8124）となる．
そのため，エントロピーの初期値は以下の通り，0.999であることがわかる．

```python
entropy_init = - (0.518 * np.log2(0.518) + 0.482 * np.log2(0.482))
print(f"毒キノコデータのエントロピーの初期値:{entropy_init:.3f}")
```

### 情報利得：分岐条件の有益さを測る

エントロピーは1に近いほど識別がされていない状態，0に近いほど識別がよくされている状態であった．
次に考えるべきことは，どの説明変数を分岐に用いたら不純度（キノコデータでは初期時点で0.999）をより小さくできるのかということである．
そこで押さえるべき概念が**情報利得（information gain）**である．
情報利得とは，ある変数を使ってデータ分割する際に，そのデータ分割前後でどれだけエントロピーが減少したかを表す指標である．
先程と同様，`cap_color_c`と`gill_color_b`の2つの変数を使い，どちらの変数が分岐条件として有益なのかを情報利得を用いて示す．
まず，`cap_color`が`c`であるか否かの2つのグループに分岐し，それぞれにおける毒キノコ割合を計算しエントロピーを計算する．

```python
mushroom_dummy.groupby(["cap_color_c","flg"])["flg"].count().unstack()
```

```python
p1 = 4176/(4176 + 3904)
p2 = 1 -p1
entropy_c0 = -(p1*np.log2(p1)+p2*np.log2(p2))
print("entropy_c0: {:.3f}".format(entropy_c0))
```

```python
p = 32/(32+12)
p2 = 1 - p1
entropy_c1 =  -(p1*np.log2(p1)+p2*np.log2(p2))
print(f"entropy_c1 {entropy_c1:.3f}")
```

分割する前の全体のエントロピーは0.999であった．
ここで分割する前のデータを親データセット，分割したデータを子のデータセットと呼ぶとした場合，情報利得を以下のように定義する．

\begin{align}
\text{情報利得} &=  \left( \text{親データセットのエントロピー} \right) - \left( \sum \frac{\text{子データセットのサイズ}}{\text{親データセットのサイズ}}\times \text{子データセットのエントロピー} \right)
\end{align}

情報利得が大きければ大きいほど，分割前後でエントロピーの低下が大きい．
そのため，より有益な分岐条件であるとわかる．
実際に$\left( \sum \frac{\text{子データセットのサイズ}}{\text{親データセットのサイズ}}\times \text{子データセットのエントロピー} \right)$の部分を計算すると，次のようになる．

```python
entropy_after = (4176 + 3904) / 8124 * entropy_c0 + (32+12)/8124*entropy_c1
print(f"データ分割後の平均エントロピー:{entropy_after:.3f}")
```

この結果，情報利得はデータ分割前後のエントロピーの差として，以下の通り0.001であることが確認できる．
つまりほとんどエントロピーが減少していないことがわかる．
このようにして，cap_colorがcかどうかは，毒キノコかどうかを見極めるために，それほど有益な分岐条件ではなさそうであるということを定量的に表現できる．

```python
print(f"変数cap_colorの分割によって得られる情報利得: {(entropy_init - entropy_after):.3f}")
```

一方，`gill_color`が`b`であるかどうかの情報利得を計算すると，以下の通り0.269となる．
上記の分岐条件よりもエントロピーを大きく低下させられるため，より有益な分岐条件とわかる．

```python
p1 = 4208 /(4208 + 2188)
p2 = 1 - p1
entropy_b0 = -(p1*np.log2(p1) + p2*np.log2(p2))

p1 = 0/(0+1728)
p2 = 1 - p1 
entropy_b1 = -(p2*np.log2(p2))
entropy_after = (4208 + 2188) / 8124 * entropy_b0 + (0+1728)/8124 * entropy_b1
print(f"変数gill_colorの分割によって得られる情報利得:{(entropy_init - entropy_after):.3f}")
```

留意すべきは，`gill_color`が`b`である場合のエントロピー計算である．
エントロピーの定義は，厳密には空ではないカテゴリについて計算するという条件がある．
`gill_color`が`b`である場合，`flg`変数が0となるサンプルはない．
そのため，エントロピー計算の$\Sigma$に$p1 \times \log_2p1$を含めていない．

```python
mushroom_dummy.groupby(["gill_color_b","flg"])["flg"].count().unstack()
```

以上で決定木の生成プロセス（条件分岐の優劣の決め方）を確認した．
情報利得が一番大きい分岐条件でデータを分割し，更に分割先でも同様に情報利得を最大とする分岐条件を探索してくれるのが決定木の動作イメージである．
ここでは，不純度を表す指標としてエントロピーを紹介した．
他にも紹介に留めるが，**ジニ不純度（Gini impurity），分類誤差（classification error）**などがある．

### 決定木のモデル構築

決定木の動きを理解したところで，決定木のモデル構築をしよう．
`sklearn.tree`モジュールの`DecisionTreeClassifier`クラスを使うことで，決定木モデルを構築できる．

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X = mushroom_dummy.drop("flg",axis=1)
y = mushroom_dummy["flg"]
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

下記のプログラムでは`DescriptionTreeClasifier`クラスを使う際，パラメータの`criteroin`に'`entropy`'を指定することで，分岐条件の指標としてエントロピーを設定している．

```python
model_en = DecisionTreeClassifier(criterion="entropy",max_depth=5,random_state=0)
model_en.fit(X_train,y_train)

print(f"正解率(train):{model_en.score(X_train,y_train):.3f}")
print(f"正解率(test):{model_en.score(X_test,y_test):.3f}")
```

下記のプログラムでは`DescriptionTreeClasifier`クラスを使う際，パラメータの`criteroin`に'`gini`'を指定することで，分岐条件の指標としてジニ係数を設定している．

```python
model_gi = DecisionTreeClassifier(criterion="gini",max_depth=5,random_state=0)
model_gi.fit(X_train,y_train)

print(f"正解率(train):{model_gi.score(X_train,y_train):.3f}")
print(f"正解率(test):{model_gi.score(X_test,y_test):.3f}")
```

結果はテストデータで89%ほどの正解率である．
決定木の分岐数決定のパラメータに`max_depth`があり，上記では5にしている．
深ければ当然，条件分岐数の上限も増える．
正解率を高めるべくより複雑なモデルにしたい場合は深い木を作ればよい．
ただし，あまり深い木を作ると過学習の危険性が増すので注意が必要である．
また決定木は，モデルを構築する際に他のモデルでは必須となる標準化処理をしなくても結果は変わらない．

## k-近傍法（k-NN）

本節では，**k-近傍法（k-Nearest Neighbor：k-NN）**について学ぶ．
たとえば，あるグループAとグループBがあり，グループ要素の属性がわかっているとして，どちらのグループに属するか分からない新しい要素を分類する場合を考える．

k-近傍法では，新しい要素と属性が近いk個の要素を選び，その要素がどちらのグループに属している割合が多いのかを調べ，多い方を新しい要素が属するグループにする．
k-近傍法のkは決定に利用する要素数に相当する．

k-近傍法はlazy learningやmemory-based learningとも言われ，訓練データをそのまま覚えておき，推論時に予測結果が計算される．
実質的な学習が推論時まで持ち越されるため，遅延学習の一種でもある．

以下は，$k$-近傍法のイメージである．
四角形をグループ$A$，星型をグループ$B$として，
三角形をどちらかのグループに分類する場合を示す．
$k=3$の場合，三角形から最も近い3つの要素すべてがグループBであるため，三角形はグループB（星型）に属すると判断される．
kを増やし$k=7$とした場合は，グループAが4つ，グループBが3つである．したがってこの場合であれば，三角形はグループA（四角形）に属すると判断される．

k近傍法は，このようにkの値によって結果が変わるため，注意が必要である．
なおk-近傍法は，マーケティングの世界ではLook-Alikeモデルとも言われ，属性が似ている人たちを集めて判断して，それぞれの属性に合ったアプローチを仕掛けていく際に活用される．
k-近傍法は回帰にも分類にも使うことができる．

<img width="380px" src="fig/k-近傍法.svg" alt="k-近傍法のイメージ">

### k-NNのモデル構築

k-NNを使ってモデル構築をしてみよう．
`sklearn.neighbors`モジュールの`KNeighborsClassifier`クラスを使う．
データ例としては乳がんに関するデータセットを使う．
乳がんに関するデータセットは`load_breast_cancer`関数で取得できる．



```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
```

```python
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=0)

training_accuracy = []
test_accuracy = []
```

ここでは`k`を1から20まで変化させ，訓練データとテストデータの正解率の変化を計算している．

```python
for i in range(1,21):
    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(X_train,y_train)
    training_accuracy.append(model.score(X_train,y_train))
    test_accuracy.append(model.score(X_test,y_test))
```

```python
plt.plot(range(1,21),training_accuracy,label="Training")
plt.plot(range(1,21),test_accuracy, label="Test")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.savefig("k-NNex.pdf")
```

`k`が小さい時は正解率に乖離があるが，`k`が6から8あたりで訓練とテストの正解率が近くなっていく．
そしてそれ以上増やしてもモデル精度に大きな変化は見られなくなる．
精度に改善が見られない場合，あまり`k`を大きくする必要はない．
そのため，本ケースにおいては`k`を6から8程度に設定しておくのが良さそうである．
なお，以下は分類タスクにおけるモデル構築の例であるが，回帰の場合は`KNeighborsRegressor`クラスを使う．

## サポートベクターマシン

**サポートベクターマシン（Support Vector Machine：SVM）**は，カテゴリを識別する境界線を，マージンが最大になるように引く手法である．
たとえば以下のような2つのグループを分ける境界線を引くとする．
このときの線の引き方は様々考えられるが，それぞれのグループの中で最も境界線に近い点（サポートベクター）との距離（マージン）が最大化するように線を引くのがサポートベクターマシンである．
厳密には，元データを更なる高次元空間へ変換した後で境界線を学習するが，時間の都合上詳細は割愛する．

<img width="380px" src="fig/サポートベクターマシン.svg" alt="サポートベクターマシンのイメージ">

### サポートベクターマシンのモデル構築

サポートベクターマシンを使ってモデルを構築してみよう．
サポートベクターマシンは`sklearn.svm`モジュールの`LinearSVC`クラスを使う．
ここではデータ例として，k-NNのモデル構築で使ったのと同じ乳がんに関するデータセットを使う．

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=0)
```

```python
model = LinearSVC()
model.fit(X_train,y_train)

print(f"正解率(train):{model.score(X_train,y_train):.3f}")
print(f"正解率(test):{model.score(X_test,y_test):.3f}")
```

サポートベクターマシンでは，標準化するとスコアが改善される．実際にやってみると改善していることがわかる．

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
model = LinearSVC()
model.fit(X_train_std,y_train)
```

```python
print(f"正解率(train):{model.score(X_train_std,y_train):.3f}")
print(f"正解率(test):{model.score(X_test_std,y_test):.3f}")
```

## 問題3

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data'
res = requests.get(url).content

mushroom = pd.read_csv(io.StringIO (res.decode('utf-8')), header=None) 
mushroom.to_csv('iagaricus-lepiota.csv', index=False, header=None) 
mushroom.columns = ['classes', 'cap_shape', 'cap_surface', 
'cap_color', 'odor','bruises','gill_attachment', 'gill_spacing', 'gill_size', 'gill_color', 'stalk_shape', 'stalk_root', 'stalk_surface_above_ring', 'stalk_surface_below_ring', 'stalk_color_above_ring', 'stalk_color_below_ring', 'veil_type', 'veil_color', 'ring_number','ring_type', 'spore_print_color', 'population', 'habitat']
mushroom.head()
mushroom_dummy = pd.get_dummies (mushroom[['gill_color', 'gill_attachment','odor','cap_color']])
mushroom_dummy['flg'] = mushroom['classes'].map (lambda x: 1 if x =='p' else 0)
mushroom_dummy = mushroom_dummy.drop(['gill_color_e', 'gill_color_h',
       'gill_color_k', 'gill_color_n', 'gill_color_o', 'gill_color_p'
       , 'gill_color_u', 'gill_color_w', 'gill_color_y',
       'gill_attachment_a', 'gill_attachment_f', 'odor_f', 'odor_t',
        'cap_color_c', 'cap_color_e',
       'cap_color_n', 'cap_color_p', 'cap_color_u',
       'cap_color_w', 'cap_color_y'],axis=1)

X = mushroom_dummy.drop("flg",axis=1)
y = mushroom_dummy["flg"]
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)

model_en = DecisionTreeClassifier(criterion="entropy",max_depth=5,random_state=0)
model_en.fit(X_train,y_train)
model_gini = DecisionTreeClassifier(criterion="gini",max_depth=5,random_state=0)
model_gini.fit(X_train,y_train)
print(f"正解率(train):{model_en.score(X_train,y_train):.3f}")
print(f"正解率(test):{model_en.score(X_test,y_test):.3f}")
print(f"正解率(train):{model_gini.score(X_train,y_train):.3f}")
print(f"正解率(test):{model_gini.score(X_test,y_test):.3f}")
```
